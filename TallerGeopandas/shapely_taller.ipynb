```python
from shapely.geometry import Point
```

```python
point2 = Point(0,0)
point3 = Point(1,2)
point4 = Point(2,4)
point5 = Point(1,9)
```

```python
from shapely.geometry import LineString
```

```python
linea1 = LineString([point2,point3,point4,point5])
linea1
```

```python
x_linea, y_linea = linea1.xy
```

```python
plt.plot(x_linea, y_linea, color='blue')
plt.grid(True)
plt.show()
```

```python
# Poligonos
from shapely.geometry import Point, Polygon
```

```python
polygono1 = Polygon([p.x, p.y] for p in [point2,point3,point4,point5])
polygono1
```

```python
x_polygono, y_polygono = polygono1.exterior.xy
```

```python
plt.fill(x_polygono, y_polygono,alpha=0.5, fc='yellow', ec='red')
plt.grid()
plt.show()
```

```python
point6 = Point(1, 4)
polygono1 = Polygon([p.x, p.y] for p in [point2,point3,point4,point5,point6])
polygono1
```

```python
x_polygono, y_polygono = polygono1.exterior.xy
plt.fill(x_polygono, y_polygono,alpha=0.5, fc='yellow', ec='red')
plt.grid()
plt.show()
```

```python
exterior = [(0,0),(3,0),(3,3),(1,3)]
hueco = [[(1,1),(2,1),(2,3),(1,2),(1,1)]]
polygon2 = Polygon(shell=exterior, holes=hueco)
polygon2
```

```python
x_ext, y_ext = polygon2.exterior.xy
plt.fill(x_ext, y_ext, alpha=1.0, fc='blue', ec='red')

for interior in polygon2.interiors:
    x_inte, y_inte = interior.xy
    plt.fill(x_inte, y_inte, alpha=1.0, fc='white', ec='black')
    
plt.show()
```

```python
from shapely.geometry import Polygon

exterior = Polygon([(0,0),(3,1),(0,3)])
hole = [[(0, 1),(3, 1),(0, 1.5)]]
polygon3 = Polygon(shell=exterior.exterior, holes=hole)
polygon3
```

```python
x_ext, y_ext = polygon3.exterior.xy
plt.fill(x_ext, y_ext, alpha=1.0, fc='red', ec='black')

for interior in polygon3.interiors:
    x_inte, y_inte = interior.xy
    plt.fill(x_inte, y_inte, alpha=1.0, fc='white', ec='black')

plt.grid()   
plt.show()
```

```python
from shapely.geometry import Polygon

# primero definimos nuestro exterior
exterior = Polygon([(0, 0), (9, 0), (9,9), (0, 9)])

# A continuacin definimos la forma del agujero como una lista de puntos
hole = [[(1, 1), (8, 1), (8, 8), (1, 8)]]

# Ahora se puede construir un polgono con los contornos y el agujero definidos
polygon5 = Polygon(shell=exterior.exterior, holes=hole)
polygon5
```

```python
x_ext, y_ext = polygon5.exterior.xy
plt.fill(x_ext, y_ext, alpha=1.0, fc='green', ec='red')

for interior in polygon5.interiors:
    x_inte, y_inte = interior.xy
    plt.fill(x_inte, y_inte, alpha=1.0, fc='white', ec='black')

plt.grid()   
plt.show()
```

```python
from shapely.geometry import Point, Polygon, LineString

exterior = Polygon([(0, 0), (10, 0), (10,10), (0, 10)])
cuadrado = Polygon([(10, 0), (20, 0), (20,10), (10, 10)])
exterior = exterior.union(cuadrado)
hole = [[(1, 1), (9, 1), (9, 9), (1, 9)]]
polygon6 = Polygon(shell=exterior.exterior, holes=hole)
polygon6
```

```python
x_ext, y_ext = polygon6.exterior.xy
plt.fill(x_ext, y_ext, alpha=1.0, fc='red', ec='black')

for interior in polygon6.interiors:
    x_inte, y_inte = interior.xy
    plt.fill(x_inte, y_inte, alpha=1.0, fc='white', ec='black')

#plt.grid()   
linea = LineString([Point(10,0),(10,10)])
x_linea, y_linea = linea.xy
plt.plot(x_linea,y_linea, color='black')
plt.show()
```
